```python
# merging the trk files using nipype
import nipype.interfaces.diffusion_toolkit as dtk
import os, sys
from tqdm import tqdm
trk_merge = dtk.TrackMerge()

path_with_trk_files = "/media/ang/Data/dMRI_data/105HCP/Fiber_Tracts/"

path_list = list()
sys.path.append(path_with_trk_files)


for i in tqdm(os.listdir(path_with_trk_files)):
    path_list = list()
    for j in os.listdir(os.path.join(path_with_trk_files,i,"tracts/")):
        # print(os.path.join(path_with_trk_files,i,"tracts", j))
        path_list.append(os.path.join(path_with_trk_files,i,"tracts", j))
    sys.path.append(os.path.join(path_with_trk_files,i))
    trk_merge.inputs.track_files = path_list
    trk_merge.inputs.output_file = os.path.join(path_with_trk_files,i, str(i)+".trk")
    trk_merge.run()
```
